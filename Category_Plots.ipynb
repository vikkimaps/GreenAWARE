```python
import pandas as pd #importing pandas
pd.options.mode.chained_assignment = None #this just turns off some error messages we don't need
import matplotlib.pyplot as plt #import matplotlib's pyplot interface, and give it the nickname plt
import seaborn as sns #for some smarter visualisations, nicknamed sns
```

```python
data = pd.read_excel("plots.xlsx")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Health'] = data['Health'].astype("category")
```

```python
data['Health'] = data['Health'].cat.set_categories(['Very Bad', 'Bad', 'Fair', 'Good', 'Very Good'], ordered=True)
```

```python
data['Frequency'] = data['Frequency'].astype("category")
```

```python
data['Frequency'] = data['Frequency'].cat.set_categories(['Every Day', '> Twice a week', 'Twice a week', 'Once a week', 'Once or twice a month', 'Once every 2-3 months', 'Less often', 'Never'])
```

```python
data['Finances'] = data['Finances'].astype("category")
```

```python
data['Finances'] = data['Finances'].cat.set_categories(['Financial difficulty', 'Struggling', 'Getting by', 'Doing ok', 'Living comfortably'])
```

```python
pal = sns.diverging_palette(145, 300, s=60)
```

```python
sns.displot(data=data, x="Frequency", hue="Health", multiple="dodge", height = 7, aspect =2.5, palette = pal)
```

```python
sns.displot(data=data, x="Frequency", hue="Finances", multiple="dodge", height = 7, aspect =2.5, palette = pal)
```
